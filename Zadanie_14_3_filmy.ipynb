```python
import pandas as pd
df_mov = pd.read_csv('tmdb_movies.csv')
df_gen = pd.read_csv('tmdb_genres.csv')
```

3. kwartyl rozkładu głosów wynosi 737, top 10 najwyżej ocenianych filmów znajduje się na powyższej tabeli.

```python
df_mov['release_date'] = pd.to_datetime(df_mov['release_date'])
df_financials = df_mov[(df_mov['release_date'] > '2010/01/01') & (df_mov['release_date'] < '2016/12/31')].groupby(pd.Grouper(key='release_date',freq='Y')).mean()
df_financials
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
def million(x, pos):
        return 'PLN {:2.1f}M'.format(x*1e-6)
x = df_financials.index
y = df_financials['budget']
z = df_financials['revenue']
axes.set_title('Średni przychód i budżety filmów w latach 2010-2016')
axes.plot(x,y, color="red")
axes.bar(x,z, color="blue")
formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)
plt.show()
```
